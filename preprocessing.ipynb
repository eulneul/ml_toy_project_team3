```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("C:/Users/Kimchiyeon/Desktop/khuda/a\새 폴더/fi.csv")
```

```python
df.info()
```

```python
idx = df[df['HSHLD_INCOME_DGREE_NM']=='모름'].index
df.drop(idx, inplace=True)
```

```python
encoding_columns = ['SEXDSTN_FLAG_CD', 'AGRDE_FLAG_NM','ANSWRR_OC_AREA_NM', 'HSHLD_INCOME_DGREE_NM',
                    'HSHLD_INCOME_DGREE_NM', 'SU_TOUR_INTRST_VALUE',
                    'GG_IC_TOUR_INTRST_VALUE','BS_TOUR_INTRST_VALUE','CHN_DJ_TOUR_INTRST_VALUE','CB_TOUR_INTRST_VALUE','GN_US_TOUR_INTRST_VALUE','GB_DG_TOUR_INTRST_VALUE','JN_GJ_TOUR_INTRST_VALUE','JNBK_TOUR_INTRST_VALUE','GW_TOUR_INTRST_VALUE','JJ_TOUR_INTRST_VALUE'

                   ]
not_encoding_columns = ['RESPOND_ID', 'EXAMIN_BEGIN_DE']
```

```python
enc_classes = {}

def encoding_label(x):
    le = LabelEncoder()
    le.fit(x)
    label = le.transform(x)
    
    enc_classes[x.name] = le.classes_
    
    return label
```

```python
d1 = df[encoding_columns].apply(encoding_label) 
d2 = df[not_encoding_columns]
d2 = d2.drop(labels='RESPOND_ID', axis=1)
data = d2.join(d1)
data.head()
```

```python
data.info()
```
