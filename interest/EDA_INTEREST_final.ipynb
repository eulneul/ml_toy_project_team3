# **카이제곱 검증**

```python
from scipy.stats import chi2_contingency
```

```python
data_drop = data1.drop(columns =['Unnamed: 0','RESPOND_ID']) #불필요한 열 제거
```

```python
data_drop
```

```python
#카이제곱 검증표를 for문으로 간편하게 만들기

c = data_drop.columns
chilist = [[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[]]
plist =  [[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[]]
for i in range(16):
    for j in range(16):
        temcrosstab = pd.crosstab(data_drop[c[i]], data_drop[c[j]], margins=False)
        result=chi2_contingency(observed=temcrosstab, correction=False)
        chilist[i].append(result[0])
        plist[i].append(result[1])
        print(result[1])
```

```python
chilist_df = pd.DataFrame(chilist)
plist_df = pd.DataFrame(plist)
```

```python
chilist_df.columns = ['날짜', '성별', '나이', '거주지', '수입', '서울', '경기', '부산', '충남', '충북', '경남', '경북', '전남', '전북', '강원', '제주']
chilist_df.index = ['날짜', '성별', '나이', '거주지', '수입', '서울', '경기', '부산', '충남', '충북', '경남', '경북', '전남', '전북', '강원', '제주']
```

# **Crame V**

```python
from scipy.stats import chi2_contingency

vlist = [[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[]]
c = data_drop.columns
for i in range(16):
    for j in range(16):
        df_cross_table = pd.crosstab(data_drop[c[i]], data_drop[c[j]])
        rows = [row.to_list() for k, row in df_cross_table.iterrows()]
        x2 = chi2_contingency(rows, correction=False)[0]
        n = np.sum(rows)
        minDimension = min(np.array(rows).shape)-1
        V = np.sqrt((x2/n) / minDimension)
        vlist[i].append(V)
```

```python
vlist_df = pd.DataFrame(vlist)
```

```python
vlist_df.columns = ['날짜', '성별', '나이', '거주지', '수입', '서울', '경기', '부산', '충남', '충북', '경남', '경북', '전남', '전북', '강원', '제주']
vlist_df.index = ['날짜', '성별', '나이', '거주지', '수입', '서울', '경기', '부산', '충남', '충북', '경남', '경북', '전남', '전북', '강원', '제주']
```
