#카이제곱 교차 검증

```python
data_drop = result1.copy()
data_drop.drop(columns = ['Unnamed: 0', 'EXAMIN_YM', 'DMSTC_TOUR_TY_VALUE', 'RESPOND_ID', 'OVSEA_TOUR_TY_VALUE'], inplace = True)
```

```python
data_drop
```

```python
#카이제곱 검증표를 for문으로 간편하게 만들기
from scipy.stats import chi2_contingency

c = data_drop.columns
chilist = [[],[],[],[]]
plist =  [[],[],[],[]]
for i in range(4):
    for j in range(4):
        temcrosstab = pd.crosstab(data_drop[c[i]], data_drop[c[j]], margins=False)
        result=chi2_contingency(observed=temcrosstab, correction=False)
        chilist[i].append(result[0])
        plist[i].append(result[1])

chilist_df = pd.DataFrame(chilist)
plist_df = pd.DataFrame(plist)

chilist_df.columns = ['성별', '나이', '거주지', '수입']
chilist_df.index = ['성별', '나이', '거주지', '수입']
plist_df.columns = ['성별', '나이', '거주지', '수입']
plist_df.index = ['성별', '나이', '거주지', '수입']

chilist_df
```

```python
import seaborn as sns

fig, ax = plt.subplots(1,1, figsize=(7, 5))
sns.heatmap(chilist_df, vmax = 10000, annot=True, fmt='.5f', cmap = 'YlGn' )
plt.title('카이제곱 검정', fontsize=20)

plt.show()
```

# **p-value**

```python
import seaborn as sns

fig, ax = plt.subplots(1,1, figsize=(7, 5))
sns.heatmap(plist_df, annot=True, fmt='.5f', cmap = 'YlGn' )
plt.title('p-value', fontsize=20)

plt.show()
```

# **Crame V**

```python
from scipy.stats import chi2_contingency

vlist = [[],[],[],[]]
c = data_drop.columns
for i in range(4):
    for j in range(4):
        df_cross_table = pd.crosstab(data_drop[c[i]], data_drop[c[j]])
        rows = [row.to_list() for k, row in df_cross_table.iterrows()]
        x2 = chi2_contingency(rows, correction=False)[0]
        n = np.sum(rows)
        minDimension = min(np.array(rows).shape)-1
        V = np.sqrt((x2/n) / minDimension)
        vlist[i].append(V)
```

```python
vlist_df = pd.DataFrame(vlist)
vlist_df.columns = ['성별', '나이', '거주지', '수입']
vlist_df.index = ['성별', '나이', '거주지', '수입']
```

```python
fig, ax = plt.subplots(1,1, figsize=(5, 5))
sns.heatmap(vlist_df, cmap = 'YlGn', annot = True, fmt = '.3f')
plt.title('Cramer V', fontsize=20)
```
